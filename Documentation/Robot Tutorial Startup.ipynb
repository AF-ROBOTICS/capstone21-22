### Powering On Robots

1. Inspect robot condition. Wheels fully seated on axles, motors fully depressed into housing, wires out of the way of wheel wells. Check battery charge on 10,000 MAh battery pack (above 30%). Check charge of AA-Batteries underneath Chassis (Must be changed weekly).
<br />

2. Ensure all robots have the proper pin connections (Red, Brown, Yellow, Orange to pins 4, 6, 8, 10 respectively).
> ![](IO_Bus.jpg)
<br />

3. Plug wire into USB-A port J12 into Micro-USB of TI-RSLK Microcontroller. 
> ![](MicroController_data2.jpg)
> ![](MicroController_Data.jpg)
<br />

4. Ensure all robots are oriented to the **west wall** to properly calibrate IMU's.
> ![](ww.jpg)
> ![](ww2.jpg)
<br />

5. Turn all robots on first by plugging USB-A end of wire into "out" of battery back located on the USAFAbot and plugging the micro-usb end of wire into "power" (J1 PWN IN) of Raspberry pi. Check for blinking green ACT LED on pi.
> ![](Pi_Power.jpg)
> ![](Pi_Power2.jpg)
<br />

6. Press button on bottom left side of TI-RSLK board. The OLED display should illuminate and a green indicator light should be seen on the board when viewed from the top of the robot. INSERT PICTURES
> ![](TI_GR.jpg)
<br />

### Ensuring Robot Connectivity
1. **DO NOT STOP ROSCORE**
<br />

2. Open New Terminal <kbd>Ctrl</kbd>+<kbd>Alt</kbd>+<kbd>t</kbd>
<br />

3. Run rostopic list. 
>![](rostopic_list2.png)

<br />

4. Ensure that one instance of */rosout, /rosout_agg, /tf, and tf_static/* are running. Then, each robot (usafabot0-24) should have */usafabot**/cmd_vel, */curr_pos, */dest_pos, */joint_states, */odom/*, and */wheel_speeds*. It may take up to 2 minutes for all the robots to appear on the list.
>![](rostopic_list.png)
<br />

5. If a robot is missing, repeat steps 4-6. Ensure that the robot is fully powered off before plugging in pi and turning on TI-RSLK board again.
<br />

6. Run rosnode list.
>![](rosnode_list2.png)
<br />

7. Ensure */rosout* is running. Each robot (usafabot0-24) after running rosnode list has the following nodes, *usafabot**/bot_to_map, */controller, */roadrunner, */robot_state_publisher, */serial*.
>![](rosnode_list.png)
<br />

8. If a robot is missing, repeat steps 4-6. Ensure that the robot is fully powered off before plugging in pi and turning on TI-RSLK board again.
<br />
